#### Sentence Transformers on HuggingFace
* `HuggingFace sentence-transformers` is a `Python framework` for `state-of-the-art sentence`, `text and image embeddings`. 
* One of the embedding models is used in the HuggingFaceEmbeddings class. 
* We have also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text="The quick brown fox jumps over the lazy dog."
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "The quick brown fox jumps over the fast dog."])
doc_result[0]
```
