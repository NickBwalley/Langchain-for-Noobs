#### Ollama
* Ollama `supports embedding models`, making it possible to `build retrieval augmented generation (RAG) applications` that `combine text prompts with existing documents` or other data.

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="llama3.2:latest")  ##by default it ues llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
       "Alpha is the first letter of Greek alphabet",
       "Beta is the second letter of Greek alphabet", 
       "Gamma is the third letter of Greek alphabet",
       "Delta is the fourth letter of Greek alphabet",
       "Epsilon is the fifth letter of Greek alphabet",
       "Zeta is the sixth letter of Greek alphabet",
       "Eta is the seventh letter of Greek alphabet",
       "Theta is the eighth letter of Greek alphabet",
       "Iota is the ninth letter of Greek alphabet",
       "Kappa is the tenth letter of Greek alphabet",
       "Lambda is the eleventh letter of Greek alphabet",
       "Mu is the twelfth letter of Greek alphabet",
       "Nu is the thirteenth letter of Greek alphabet",
       "Xi is the fourteenth letter of Greek alphabet",
       "Omicron is the fifteenth letter of Greek alphabet",
       "Pi is the sixteenth letter of Greek alphabet",
       "Rho is the seventeenth letter of Greek alphabet",
       "Sigma is the eighteenth letter of Greek alphabet",
       "Tau is the nineteenth letter of Greek alphabt",
       "Upsilon is the twentieth letter of Greek alphabet",
       "Phi is the twenty-first letter of Greek alphabet",
    ]
)
```
